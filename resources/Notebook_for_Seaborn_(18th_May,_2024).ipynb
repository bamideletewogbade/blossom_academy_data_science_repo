# **Seaborn**

#### Seaborn is a powerful Python data visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset("tips")
```

```python
# Display the first few rows of the dataframe
df.head()
```

## 1. **Pairplot**
A pairplot is useful for visualizing the distribution of data and relationships between variables.

```python
# Pairplot
sns.pairplot(df)
plt.show()
```

```python
# Pairplot
sns.pairplot(df,hue='sex')
plt.show()
```

```python
# Pairplot
sns.pairplot(df,hue='smoker',palette=['yellow','red'])
plt.show()
```

## **2. Dist Plot**
Dist plot helps us to check the distribution of the columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins =10)
```

## **3. Heatmap**
A heatmap displays data in a matrix format, using color to represent values. It's often used for correlation matrices.

```python
# Compute the correlation matrix
numeric_df = df.select_dtypes(include=[np.number])
corr = numeric_df.corr()

# Heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```

## **4. Boxplot**
A boxplot displays the distribution of data based on a five-number summary (minimum, first quartile, median, third quartile, and maximum).

```python
# Boxplot
sns.boxplot(x='sex', y='total_bill', data=df,palette=['yellow', 'red'])
plt.show()
```

```python
# Boxplot
sns.boxplot(x='smoker', y='total_bill', data=df,palette=['green', 'red'])
plt.show()
```

```python
# Boxplot
sns.boxplot(x='day', y='total_bill', data=df,palette=['yellow', 'red','green','blue'])
plt.show()
```

```python
# Boxplot
sns.boxplot(x='day', y='total_bill', data=df,palette='rainbow')
plt.show()
```

```python
# Boxplot
sns.boxplot(data=df,palette='rainbow')
plt.show()
```

```python
# Boxplot
sns.boxplot(x='day', y='total_bill',hue='smoker', data=df,palette=['yellow', 'red'])
plt.show()
```

```python
# Boxplot
sns.boxplot(x='day', y='total_bill',hue='sex', data=df,palette='rainbow')
plt.show()
```

## **5. Bar Plot**
A bar plot is a plot that represents data with rectangular bars where the length of each bar is proportional to the value it represents.

```python
# Bar plot
sns.barplot(x='sex', y='total_bill', data=df,palette=['yellow', 'green'])
plt.show()
```

```python
# Bar plot
sns.barplot(x='smoker', y='total_bill', data=df,palette=['blue', 'red'])
plt.show()
```

# **6. Joint Plot**
A joint plot allows to study the relationship between 2 numeric variables. The central chart display their correlation, it is usually a hexbin plot, a 2D histogram or a 2D density plot.

```python
# Joint plot
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
plt.show()
```

## **7. Count Plot**
A countplot is a type of bar plot that displays the counts of observations in each categorical bin using bars. It is particularly useful for visualizing the distribution of a categorical variable.

```python
# Countplot
sns.countplot(x='sex', data=df,palette=['purple', 'orange'])
plt.show()
```

```python
# Countplot
sns.countplot(x='smoker', data=df,palette=['red', 'blue'])
plt.show()
```

```python
# Countplot
sns.countplot(x='day', data=df,palette=['yellow', 'red','green','blue'])
plt.show()
```

```python
# Countplot
sns.countplot(y='sex', data=df,palette=['blue', 'red'],)
plt.show()
```
